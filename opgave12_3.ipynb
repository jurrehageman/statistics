# Test non-parametric statistics

```python
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
```

```python
males = [3, 2, 1, 4, 3, 2, 9, 13, 11, 3, 18, 2, 4, 6, 2, 1, 8, 5, 1, 14]
females = [6, 5, 2, 1, 7, 2, 2, 11, 2, 1, 1, 3, 11, 3, 2, 3, 2, 1, 4, 8]
```

```python
data = list(zip(males, females))
df = pd.DataFrame(data, columns=['males','females'])
display(df)
```

```python
tidy_df = pd.melt(df)
tidy_df = tidy_df.rename(columns={'variable': 'sex', 'value': 'protein_conc'})
display(tidy_df)
```

```python
sns.boxplot(x = 'sex', y = 'protein_conc', data = tidy_df)
sns.set_style("whitegrid")
```

## Rank-Sum

```python
res_ranksums = stats.ranksums(df['males'], df['females'])
print(res_ranksums)
```

## Mann-Whitney U

```python
res_mann_whitney = stats.mannwhitneyu(df['males'], df['females'])
print(res_mann_whitney)
```
